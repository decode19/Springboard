# Merging Columns

```python
close=pd.DataFrame({'Anew':A.Close, 'Gnew': G.Close})
close.head()
```

### Note that in the above we are creating a new dataframe called"close" and it has columns from A and G

# Cumulative sum and rolling calculations 

```python
A_change = np.log(A.Close/A.Close.shift(1))
G_change = np.log(G.Close/G.Close.shift(1))
```

```python
A_change.cumsum().plot()
G_change.cumsum().plot()
```

# Creating new dataframe and exporting it to a csv file

```python
new=pd.DataFrame({"Anew": A_change, "Gnew": G_change})
new.head()
```

#### Note: this gives a new dataframe. the first row is empty because it does not have values to calculate, so we will try to eliminate the first row in next command.

```python
new=pd.DataFrame({"Anew": A_change[1:], "Gnew": G_change[1:]})
new.head()
```

```python
new.to_csv("stocks.csv")
pd.read_csv("stocks.csv").head()
```

#### Note the first one converts to csv and the second one converts back from csv to dataframe
